```python
r = requests.get('https://www.qidian.com/rank/fin?dateType=3&chn=-1&page=1')
r.text
```

```python
from bs4 import BeautifulSoup
mypage = BeautifulSoup(r.text)
```

```python
book_writer = mypage.find_all('div',attrs={'class':'book-mid-info'})
mylist=[]
for i in range(0,20):
    mylist.append([book_writer[i].find('h4').text.strip(),book_writer[i].find('a',attrs={'class':'name'}).text.strip()])
```

```python
import csv
with open('yanyang.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)

    spamwriter.writerows(mylist)
       
```
